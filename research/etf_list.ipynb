# 获取当前所有上市 ETF 列表
本 notebook 会从 `config.toml` 读取 Tushare token，然后执行 `pro.etf_basic(list_status='L', ...)`。

```python
from pathlib import Path
import pandas as pd
import tushare as ts

try:
    import tomllib  # Python 3.11+
except ModuleNotFoundError:
    import tomli as tomllib  # Python <3.11
```

```python
project_root = Path.cwd().resolve().parent if Path.cwd().name == 'research' else Path.cwd().resolve()
config_path = project_root / 'config.toml'

with config_path.open('rb') as f:
    cfg = tomllib.load(f)

token = cfg['tushare']['token']
ts.set_token(token)
pro = ts.pro_api()

df = pro.etf_basic(
    list_status='L',
    fields='ts_code,extname,index_code,index_name,exchange,mgr_name'
)

print(f'已获取上市 ETF 数量: {len(df)}')
df.head(20)
```

```python
output_path = project_root / 'research' / 'etf_list_current.csv'
df.to_csv(output_path, index=False, encoding='utf-8-sig')
output_path
```
